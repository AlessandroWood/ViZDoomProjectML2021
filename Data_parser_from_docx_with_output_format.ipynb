```python
for i in range(1,44):
  idx = j + 8*(i-1) + 1
  line = document.paragraphs[idx+1].text
  line = line.split(']')[1]
  train.append(line)
  line = document.paragraphs[idx+4].text
  line = line.split(']')[1]
  test.append(line)
  line = document.paragraphs[idx+6].text
  time.append(line)
```

```python
print(len(time))
```

```python
mean_train = []
std_train = []
min_train = []
max_train = []
mean_test = []
std_test = []
min_test = []
max_test = []
```

```python
for line in train:
  mean_std = line.split(',')[0]
  min = line.split(',')[1]
  max = line.split(',')[2]
  min = min.split(':')[1]
  min = float(min.strip(' '))
  max = max.split(':')[1]
  max = float(max.strip(' '))
  mean = mean_std.split('+/-')[0]
  std = mean_std.split('+/-')[1]
  print(std)
  std = float(std.strip(' '))
  mean = mean.split('mean:')[1]
  mean = float(mean.strip(' '))
  mean_train.append(mean)
  std_train.append(std)
  min_train.append(min)
  max_train.append(max)
```

```python
print(mean_train)
```

Results: mean: -87.4 +/- 178.6, min: -380.0 max: 95.0

```python
for line in test:
  mean_std = line.split(',')[0]
  min_max = line.split(',')[1]
  min = min_max.split('max:')[0]
  min = min.split(':')[1]
  min = float(min.strip(' '))
  max = min_max.split('max:')[1]
  max = float(max.strip(' '))
  mean = mean_std.split('+/-')[0]
  std = mean_std.split('+/-')[1]
  print(std)
  std = float(std.strip(' '))
  mean = mean.split('mean:')[1]
  mean = float(mean.strip(' '))
  mean_test.append(mean)
  std_test.append(std)
  min_test.append(min)
  max_test.append(max)
```
