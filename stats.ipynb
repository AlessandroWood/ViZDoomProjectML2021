```python
import matplotlib.pyplot as plt
import json

with open("content/stats.json") as json_file:
    stats = json.load(json_file)

```

```python
x = range(1, stats["epochs"] + 1)
y = [e["mean"] for e in stats["train_scores"]]

plt.plot(x,y)
```

```python
x = range(1, stats["epochs"] + 1)
y = [e["mean"] for e in stats["test_scores"]]

plt.plot(x,y)
```

```python
x = range(1, stats["epochs"] + 1)
y = [e["mean"] for e in stats["loss"]]

plt.plot(x,y)

```
